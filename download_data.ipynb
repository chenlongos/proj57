# 下载模型和数据

从 HuggingFace `bobodai/proj57_model` 下载预训练模型和数据集。仅需运行一次。

```python
import os
from pathlib import Path

REPO_ID = "bobodai/proj57_model"
LOCAL_DIR = "proj57_database&model"

if Path(LOCAL_DIR).exists():
    print(f"{LOCAL_DIR}/ 已存在，跳过下载。如需重新下载请先删除该目录。")
else:
    print(f"正在从 HuggingFace 下载 {REPO_ID} ...")
    os.system(f"pip install -q huggingface_hub")
    from huggingface_hub import snapshot_download
    snapshot_download(
        REPO_ID,
        local_dir=LOCAL_DIR,
        resume_download=True,
    )
    print(f"下载完成 -> {LOCAL_DIR}/")

# 验证
print(f"\n文件数: {sum(1 for _ in Path(LOCAL_DIR).rglob('*') if _.is_file())}")
print(f"大小: {sum(f.stat().st_size for f in Path(LOCAL_DIR).rglob('*') if f.is_file()) / 1024**2:.0f} MB")
```
